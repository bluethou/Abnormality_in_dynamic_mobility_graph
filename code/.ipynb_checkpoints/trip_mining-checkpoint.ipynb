Trip mining.
 To extract travel data from Twitter, this paper creates a 3-way tensor from Twitts by reference to the research of Linus. The research team tracked a SNS user's mobility with geolocation tags of the Twitts. The user's trajectory is considered as a continuous stream of check-ins: A check-in is a tuple of the unique identifier of a user (UUID), a location, and a timestamp. They adopted a heuristic assumption that a user’s home country is the one with most check-ins. On top of that, this paper newly assumes that travelers go back to their home place on the day after the last checkin. 
 The team of Linus stored travel data with a JSON format. Additionally, it also includes intra-country trajectories.
Thus we process the data to acuiqre graphs of inter-country trips. We created multiple graphs by time intervals.

# Deriving Tourist Mobility Patterns from Check-in Data: introduce trip-mining by checks-in stream data from SNS

```python
import json
import numpy as np
import pandas as pd
import networkx as nx
```

```python
jdata = json.loads(open ('../../data/trips_europeans_pp.json').read())
# for c in jdata["trips"][3]["blocks"]:
#     print('country_code: {}, first_checkin: {}, last_checkin: {}'.format(c.get('country_code'), c.get('first_checkin'), c.get('last_checkin')))
# #     c.get('country_code')
# #     c.get('first_checkin')
# #     c.get('last_checkin')
```

```python
length = len(jdata["trips"])
df = pd.DataFrame(np.zeros(shape=(1,4)))
```

```python
ind = 1
init_trip = jdata["trips"][0]
df = pd.DataFrame([ind,init_trip["trip_id"],init_trip["traveler_home"],init_trip["first_checkin"],init_trip["last_checkin"]]).T

for trip in jdata["trips"]:
    temp_df = pd.DataFrame([ind,trip["trip_id"],trip["traveler_home"],trip["first_checkin"],trip["last_checkin"]]).T
    df = df.append(temp_df, ignore_index=True)
    for c in trip["blocks"]:
        temp_df2 = pd.DataFrame([ind,trip["trip_id"],c.get("country_code"), c.get("first_checkin"), c.get("last_checkin")]).T
        df=df.append(temp_df2, ignore_index=True)
    ind = ind+1
#     if ind > 10:
#         break
```

```python
df.columns = ["trip_number","trip_id","node","First_checkin","Last_checkin"]
```

```python
uniq_nd = df.groupby('trip_number')['node'].unique()
uniq_nd['intra_travel'] = uniq_nd.apply(lambda x: len(x) in [1])
intra_boolean=uniq_nd['intra_travel'] #this is an indicator if the trips is intra
df_intra = df.join(intra_boolean, on='trip_number', rsuffix='_intra')
df_intra
```

```python
# Get only intercountry trips
processed_df = df_intra.loc[df_intra['node_intra'] == False]
```

```python
# Save as csv
processed_df.to_csv(r'trip_data.csv', index = False)
```

```python
travel_df = processed_df
data = {'From':  ['A'], 'To': ['B'], 'Time':[0]}
from_to_df = travel_df.iloc[:,1:4]
```

```python
# Create from / to / date dataframe.
# This takes < 1hour for european travel datasets
from datetime import timedelta 
from datetime import datetime

home = travel_df.iloc[0]['node']
for i in range(1,len(travel_df)):
    if travel_df.iloc[i,0]!=travel_df.iloc[i-1,0] :
        from_to_df.iloc[i,0]= travel_df.iloc[i-1,2]
        from_to_df.iloc[i,1]= home
        from_to_df.iloc[i,2]= datetime.strptime(travel_df.iloc[i-1,4], '%Y-%m-%d %H:%M:%S') + timedelta(days=1)
        home = travel_df.iloc[i]['node']
#     if travel_df.iloc[i,0]==travel_df.iloc[i-1,0] :
    else :
        from_to_df.iloc[i,0]= travel_df.iloc[i-1,2]
        from_to_df.iloc[i,1]= travel_df.iloc[i,2]
        from_to_df.iloc[i,2]= travel_df.iloc[i,3]
```

```python
df = from_to_df.iloc[1:,].reset_index().iloc[:,1:]
df.columns = ['from','to','date']
df['date']=pd.to_datetime(df['date'])
df.to_csv(r'from_to_date.csv', index = False)
```

```python
# Initialize Graph model. 
G = nx.DiGraph()
nodes = travel_df['node'].unique().T
for u in nodes:
    for v in nodes:
        G.add_edge(u,v)
        G[u][v]['weight'] = 0
```

```python
# # You may draw a graph 
import matplotlib.pyplot as plt
G = serial_graphs[1]
edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())
pos = nx.spring_layout(G)
colors = np.asarray(weights)
colors = np.log(colors+0.001)-min(np.log(colors+0.001))
plt.figure(figsize=(12,12)) 
nx.draw(G, pos, node_color='#A0CBE2', edgelist=edges, edge_color=colors, with_labels = True, edge_cmap=plt.get_cmap('Reds'))
print("trips between",week_slices[1],"and",week_slices[2])
plt.show()
```

```python
# Create tensors whose dimension is Time x Nodes x Featrues
# Here we refer the paper of 'Anomaly detection in spatiotemporal data via regularized non-negative tensor analysis'
# The number of features = 2 x the number of Nodes
import tensorly.backend as T
time_dim # the number of slices by time which we set above.
n_nodes = travel_df['node'].unique().size
tensor = T.tensor(np.zeros(Tintervals * n_nodes * 2*n_nodes).reshape((Tintervals, n_nodes, 2*n_nodes))) # Time x Nodes x Features
```

```python
# Fill the Tensors
for i in range(0,time_dim):    
    mat = nx.adjacency_matrix(serial_graphs[i]).todense()
    feature_mat = np.hstack((mat, mat.T)) # Feature matrix.
    tensor[i] = feature_mat
```
