```python
ind = 0
init_trip = jdata["trips"][0]
df = pd.DataFrame([init_trip["traveler_home"],init_trip["first_checkin"],init_trip["last_checkin"]]).T

for trip in jdata["trips"]:
    temp_df = pd.DataFrame([trip["traveler_home"],trip["first_checkin"],trip["last_checkin"]]).T
    df = df.append(temp_df)
    for c in trip["blocks"]:
        temp_df2 = pd.DataFrame([c.get("country_code"), c.get("first_checkin"), c.get("last_checkin")]).T
        df=df.append(temp_df2)
    ind = ind+1
    if ind > 10:
        break
```
