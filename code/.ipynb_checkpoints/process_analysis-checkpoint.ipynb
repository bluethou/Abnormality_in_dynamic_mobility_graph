```python
df= travel_time_df[travel_time_df['Country']=='RU']
df['Arrival_year'] = pd.DatetimeIndex(df['Arrival_date']).year
df['Arrival_month'] = pd.DatetimeIndex(df['Arrival_date']).month
df
```

```python
year_cond1 = (df['Arrival_year']==2018)
month_cond1  = (df['Arrival_month'] == 7)| (df['Arrival_month'] == 8)
df1 = df[year_cond1 & month_cond1]

year_cond2 = (df['Arrival_year']!=2018)
month_cond2  = (df['Arrival_month'] == 7)| (df['Arrival_month'] == 8)
df2 = df[year_cond2 & month_cond2]
```

```python
event_hours = []
normal_hours = []
for i in range(len(df1)):
    event_hours.append(df1.iloc[i]['stay_time'].total_seconds()/3600)
    
for i in range(len(df2)):
    normal_hours.append(df2.iloc[i]['stay_time'].total_seconds()/3600)
event_hours = np.array(event_hours)
normal_hours = np.array(normal_hours)
# normal_hours = stats.trim_mean(normal_hours, 0.05)
# event_hours = stats.trim_mean(event_hours, 0.05)
```

```python
trim_event = event_hours[event_hours<np.percentile(event_hours,95)]
trim_normal = normal_hours[normal_hours<np.percentile(normal_hours,95)]
```

```python
plt.figure(figsize=(12,5))
plt.hist(trim_normal, bins = 30, density=True, alpha=0.7, label="June & July not in 2018")
plt.hist(trim_event, bins = 30, density=True, alpha=0.7, label="June & July in 2018")
plt.axvline(x=np.mean(trim_normal), linestyle='--')
plt.axvline(x=np.mean(trim_event), linestyle='--', color='orange')

plt.title("Density Histogram of Travel Hours")
plt.ylabel("Density")
plt.xlabel("Travel time(h)")

plt.legend()
plt.show()
```
