```python
import numpy as np
from cdlib import algorithms, viz, evaluation
import networkx as nx
import pickle
import pandas as pd
```

```python
with open('../pkl/Twitter_label_dict.pkl', 'rb') as f:
    label_dict = pickle.load(f) 
    
with open('../pkl/global_weekly_sorted_tensor2.pkl', 'rb') as f:
    travel_tensors = pickle.load(f) 
```

```python
travel_tensors.shape
```

```python
week_slices = pd.date_range('2016-01-04', periods=200, freq='W')
```

```python
week_slices
```

```python
week_slices[157]
```

```python
label_dict
```

```python
adj_list = []
coms_list = []
G_list = []
for i in range(11):
#     ajm = np.sum(travel_tensors[157 + 4*i:161 + 4*i][:,0:226,0:226], axis=0)
    ajm = np.sum(travel_tensors[156 + 4*i:160 + 4*i][:,0:100,0:100], axis=0)
    adj_list.append(ajm)
    temp_G = nx.from_numpy_matrix(ajm, create_using=nx.DiGraph)
    temp_G = nx.relabel_nodes(temp_G, label_dict) 
    G_list.append(temp_G)
    coms_list.append(algorithms.leiden(temp_G, weights='weight'))
```

```python
print(coms_list[3].communities[0:5])
```

```python
G = G_list[3]
coms = coms_list[3]
pos = nx.kamada_kawai_layout(G)
# pos = nx.spring_layout(G)
viz.plot_network_clusters(G, coms, pos, figsize=(20, 20), plot_labels=True, node_size=800)
# viz.plot_community_graph(G, coms)
```
